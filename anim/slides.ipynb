```python
import manim as mm
```

```python
%%manim -qm -v WARNING slide1

class slide1(mm.Scene):
    def construct(self):
        self.camera.background_color = "#ece6e2"
        logo_green = "#87c2a5"
        logo_blue = "#525893"
        logo_red = "#e07a5f"
        logo_black = "#343434"
        ds_m = mm.MathTex(r"\mathbb{M}", fill_color=logo_black).scale(7)
        ds_m.shift(2.25 * mm.LEFT + 1.5 * mm.UP)
        circle = mm.Circle(color=logo_green, fill_opacity=1).shift(mm.LEFT)
        square = mm.Square(color=logo_blue, fill_opacity=1).shift(mm.UP)
        triangle = mm.Triangle(color=logo_red, fill_opacity=1).shift(mm.RIGHT)
        logo = mm.VGroup(triangle, square, circle, ds_m)  # order matters
        logo.move_to(mm.ORIGIN)
        self.add(logo)

#s1 = slide1()
#s1.render(preview=True)
```

```python
%%manim -qm -v WARNING slide1

class slide1(mm.Scene):
    def construct(self):
        self.camera.background_color = "#ece6e2"
        logo_green = "#87c2a5"
        logo_blue = "#525893"
        logo_red = "#e07a5f"
        logo_black = "#343434"
        ds_m = mm.MathTex(r"{H} = \sum_{i}\left[t_0 a_i^\dagger b_i + t_1 a_i^\dagger b_{i+1} + \rm{h.c.}\right]", fill_color=logo_black).scale(1.5)
        ds_m.shift(2.25 * mm.LEFT + 1.5 * mm.UP)
        circle = mm.Circle(color=logo_green, fill_opacity=0.5).shift(mm.LEFT).scale(0.1)
        square = mm.Square(color=logo_blue, fill_opacity=0.5).shift(mm.UP).scale(0.1)
        triangle = mm.Triangle(color=logo_red, fill_opacity=0.5).shift(mm.RIGHT).scale(0.1)
        logo = mm.VGroup(triangle, square, circle, ds_m)  # order matters
        logo.move_to(mm.ORIGIN)
        self.add(logo)
```
