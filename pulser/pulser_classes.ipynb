```python
import numpy as np
import pulser
import pulser.waveforms
```

```python
duration = 1000 # ~ 1 microsec
black = pulser.waveforms.BlackmanWaveform(duration, np.pi)
ramp = pulser.waveforms.RampWaveform(duration, -5.0, 10.0)
pulse = pulser.Pulse(amplitude=black, detuning=ramp, phase=0)
pulse.draw()
```

```python
a = pulser.waveforms.KaiserWaveform(duration, 1.1)
```
