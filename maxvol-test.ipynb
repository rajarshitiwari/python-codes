```python
import numpy as np
```

```python
n, r = 100, 6

A = np.random.random((n, r))
np.linalg.matrix_rank(A)
```

```python
for i in range(n-r+1):
    Ar = A[i:i+r, :]
    B = A @ np.linalg.inv(Ar)
    x, y = np.where(abs(B) == abs(B).max())
    b1 = abs(B).max()
    if b1 > 1:
        B[[y, x]] = B[[x, y]]
        A[[y, x]] = A[[x, y]]
    b2 = abs(B).max()
    print(abs(np.linalg.det(Ar)), B.shape, x, y, b1, b2-b1)
```
