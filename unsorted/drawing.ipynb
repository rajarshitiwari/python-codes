```python
import networkx as nx
import matplotlib.pyplot as plt
L = 36
g = nx.Graph()
g.add_nodes_from([(_, {'color': 'red' if _%2==0 else 'green'}) for _ in range(L)])
# NN bonds
#g.add_edges_from([(_, (_ + 1) %L) for _ in range(L)])

for nn in range(1, L//2):
    g.add_weighted_edges_from([(_, (_ + nn) %L, 2/nn) for _ in range(L)])


#nx.draw_shell(g)
#nx.draw_networkx(g)
#nx.draw_random(g)
colors = [i['color'] for i in g.nodes.values()]
opts = {'with_labels': True, 'font_weight': 'bold', 'node_color': colors, 'edge_color': 'black'}
#nx.draw(g, **opts)
nx.draw_circular(g, **opts)
```
