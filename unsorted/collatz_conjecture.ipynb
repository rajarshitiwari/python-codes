```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 3 * x + 1 if x % 2 == 1 else x //2

def get_numlist(num, length):
    ll = np.zeros(length, dtype=int)
    x = num
    for i in range(length):
        ll[i] = x
        x = func(x)
    return ll
```

```python
ll = get_numlist(1200, 120)
```

```python
plt.plot(ll)
plt.yscale('log')
```
